```python
from bs4 import BeautifulSoup
import requests
```

```python
source = requests.get("https://www.imdb.com/search/keyword/?keywords=anime&mode=advanced&page=1&ref_=kw_nxt&sort=moviemeter,asc")
source.raise_for_status()
soup = BeautifulSoup(source.content, "html.parser")
Animelist = soup.find("div", class_ = "lister-list").find_all("div", class_ = "lister-item mode-advanced")
```

```python

Name = []
Years = []
Genre = []
Rating = []
```

```python
for anime in Animelist:
    Rating_web = anime.find("div", class_ = "inline-block ratings-imdb-rating").text.strip("\n")
    Rating.append(Rating_web)
     
    Genre_web = anime.find("span", class_ = "genre").text.strip("\n")
    Genre.append(Genre_web)
    
    Year_web  = anime.find("span", class_ ="lister-item-year text-muted unbold" ).text.strip("()")
    Years.append(Year_web)
    
    Name_web = anime.find("h3", class_ = "lister-item-header").a.text
    Name.append(Name_web)
```

```python
AnimeDF = pd.DataFrame({"Anime_Title" : Name ,"Ratings" : Rating ,  "Year_Active" : Years, "Genres" : Genre})
```

```python
AnimeDF
```

```python
AnimeDF.to_csv('Top 50 Anime of all time.csv', index=False, encoding="utf-8-sig")
```
